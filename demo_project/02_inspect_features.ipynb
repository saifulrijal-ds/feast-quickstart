```python
import duckdb
import pandas as pd
```

```python
# Connect to DuckDB
con = duckdb.connect()

# Install and load SQLite extension
con.sql("INSTALL sqlite;")
con.sql("LOAD sqlite;")

# Attach SQLite database
con.sql("ATTACH 'feature_repo/data/registry.db' AS registry_db (TYPE sqlite);")

# Query to list all tables
tables_query = """
SELECT 
    name as table_name,
    type,
    sql as create_statement
FROM registry_db.sqlite_master 
WHERE type='table'
ORDER BY table_name;
"""

# Get results as pandas DataFrame
tables_df = con.sql(tables_query).df()

# Display tables
print("Tables in the database:")
print(tables_df[['table_name', 'type']])

# If you want to see the CREATE statement for a specific table
print("\nCreate statement for first table:")
print(tables_df['create_statement'].iloc[0])
```
